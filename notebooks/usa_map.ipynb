```
from matplotlib import pyplot as plt
import json
import numpy as np

fsize=20

plt.rcParams.update({'font.size': fsize})
%config InlineBackend.figure_format = 'retina'
```

```
# data via https://eric.clst.org/tech/usgeojson/
with open('/content/gz_2010_us_040_00_500k.json', 'r') as f:
    state_json = json.load(f)
```

```
fig, ax = plt.subplots(1,1, constrained_layout=True, figsize=(11,6))

for j,state in enumerate( np.random.permutation(state_json['features']) ):
    for subregion in state['geometry']['coordinates']:
        xy = np.squeeze(subregion)
        ax.fill(xy[:,0], xy[:,1], fc='w', ec='k', )

kwd = {
    "title": 'United States of America',
    "xlim": [-130,-65],
    "ylim": [24,51],
    "xticks": [],
    "yticks": [],
}
ax.set(**kwd)

# Hide grid lines
ax.grid(False)
ax.axis("off")

fig.show()
```
